```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import fftpack

## Parametros
K = 20                                          # Número de amostras
l = np.arange(K)                                # Eixo amostras
 
# Definindo forma de onda
s_0 = np.ones(K)                                                   # Sinal 1: Um degrau
s_1 = np.concatenate((s_0[0:int(K/2)] , -s_0[int(K/2):K]))         # Sinal 2: Soma de degraus

#_1 = np.arange(1,2);    # Sinal 2: Soma de degraus

# Inicializando sinais de saida do correlator
#r_0 = np.zeros(K)                            # Inicializa vetor r_0
#r_1 = np.zeros(K)                            # Inicializa vetor r_1

vtVar = np.array((0, 0.1, 1))                          # Variâncias a simular

# Figura para sinais no tempo
fig1 = plt.figure(figsize = (15, 20))

plt.subplot(len(vtVar)*2+1, 2, 1)
plt.plot(l,s_0)
plt.legend(['s_0(t)'])
plt.xlim([0, 19])
plt.ylim([min(s_0) - 1, max(s_0) + 1])

plt.subplot(len(vtVar)*2+1, 2, 2)
plt.plot(l,s_1)
plt.legend(['s_1(t)'])
plt.xlim([0, 19])
plt.ylim([min(s_1) - 1, max(s_1) + 1])

# Figura para saídas dos correlatores
# Cálculo da saída do correlator para cada valor de variância do ruído

for  ik in range(len(vtVar)):
    vtNoise = np.sqrt(vtVar[ik])*np.random.randn(K)        # Vetor de ruído
    
    # Sinais quando s_0 é transmitido
    rs_0 = s_0 + vtNoise                                  # Sinal recebido
    
    # Correlacionando sinais
    for n in range(K):
        r_0[n] = np.sum(rs_0[0:n]*s_0[0:n])
        r_1[n] = np.sum(rs_0[0:n]*s_1[0:n])
    ## Gráficos 
    plt.subplot(len(vtVar)+4, 2, 2*(ik)+9)
    plt.plot(l,r_0)
    plt.plot(l,r_1, '--')
    plt.legend(['r_0', 'r_1'])
    plt.xlim([0, 19])
    plt.ylim([min(min(r_0), min(r_1)) - 2, max(max(r_0), max(r_1)) + 2])
    
    # Gráficos da autocorrelação
    plt.subplot(len(vtVar)*2+1, 2, 2*(ik)+3)
    plt.plot(l,rs_0)
    plt.legend(['Variância:' + str(vtVar[ik])])
    plt.xlim([0, 19])
    plt.ylim([min(rs_0) - 1, max(rs_0) + 1])
    
    
    ## Sinais quando s_1 é transmitido    
    rs_1 = s_1 + vtNoise                         # Sinal recebido
    
    # Correlacionando sinais
    for n in range(K):
        r_0[n]=np.sum(rs_1[0:n]*s_0[0:n])
        r_1[n]=np.sum(rs_1[0:n]*s_1[0:n])

    ## Gráficos 
    plt.subplot(len(vtVar)+4, 2, 2*(ik)+10)
    plt.plot(l,r_0)
    plt.plot(l,r_1,'--')
    plt.legend(['r_0', 'r_1'])
    plt.xlim([0, 19])
    plt.ylim([min(min(r_0), min(r_1)) - 2, max(max(r_0), max(r_1)) + 2])
    
    # Gráficos da autocorrelação
    plt.subplot(len(vtVar)*2+1, 2, 2*(ik)+4)
    plt.plot(l,rs_1)
    plt.xlim([0, 19])
    plt.ylim([min(rs_1) - 1, max(rs_1) + 1])
    plt.legend(['Variância:' + str(vtVar[ik])])
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import fftpack

## Parametros
K = 20                                          # Número de amostras
l = np.arange(K)                                # Eixo amostras
 
# Definindo forma de onda
s_0 = np.ones(K)                                                   # Sinal 1: Um degrau
s_1 = np.concatenate((s_0[0:int(K/2)] , -s_0[int(K/2):K]))         # Sinal 2: Soma de degraus

vtVar = np.array((0, 0.1, 1))                  # Variâncias a simular

# Figura para sinais no tempo
fig1 = plt.figure(figsize = (15, 20))

plt.subplot(len(vtVar)*2+1, 2, 1)
plt.plot(l,s_0)
plt.legend(['s_0(t)'])
plt.xlim([0, 19])
plt.ylim([min(s_0) - 1, max(s_0) + 1])

plt.subplot(len(vtVar)*2+1, 2, 2)
plt.plot(l,s_1)
plt.legend(['s_1(t)'])
plt.xlim([0, 19])
plt.ylim([min(s_1) - 1, max(s_1) + 1])

# Figura para saídas dos correlatores
# Cálculo da saída do correlator para cada valor de variância do ruído

for  ik in range(len(vtVar)):
    vtNoise = np.sqrt(vtVar[ik])*np.random.randn(K)        # Vetor de ruído
    
    # Sinais quando s_0 é transmitido
    rs_0 = s_0 + vtNoise                                  # Sinal recebido
    
    # Filtro Casado
    r_0 = np.convolve(rs_0,np.flip(s_0, 0))               # Convolução filtro casado saída 0
    r_0 = r_0[0:int(np.ceil(len(r_0)/2))]
    
    r_1 = np.convolve(rs_0,np.flip(s_1, 0))               # Convolução filtro casado saída 1
    r_1 = r_1[0:int(np.ceil(len(r_1)/2))]
     
    ## Gráficos 
    plt.subplot(len(vtVar)+4, 2, 2*(ik)+9)
    plt.plot(l,r_0)
    plt.plot(l,r_1, '--')
    plt.legend(['r_0', 'r_1'])
    plt.xlim([0, 19])
    plt.ylim([min(min(r_0), min(r_1)) - 2, max(max(r_0), max(r_1)) + 2])
    
    # Gráficos da autocorrelação
    plt.subplot(len(vtVar)*2+1, 2, 2*(ik)+3)
    plt.plot(l,rs_0)
    plt.legend(['rs_0, Variância:' + str(vtVar[ik])])
    plt.xlim([0, 19])
    plt.ylim([min(rs_0) - 1, max(rs_0) + 1])
    
    ## Sinais quando s_1 é transmitido    
    rs_1 = s_1 + vtNoise                         # Sinal recebido
    
    # Filtro Casado
    r_0 = np.convolve(rs_1,np.flip(s_0, 0))               # Convolução filtro casado saída 0
    r_0 = r_0[0:int(np.ceil(len(r_0)/2))]
    
    r_1 = np.convolve(rs_1,np.flip(s_1, 0))               # Convolução filtro casado saída 1
    r_1 = r_1[0:int(np.ceil(len(r_1)/2))]
    
    ## Gráficos 
    plt.subplot(len(vtVar)+4, 2, 2*(ik)+10)
    plt.plot(l,r_0)
    plt.plot(l,r_1,'--')
    plt.legend(['r_0', 'r_1'])
    plt.xlim([0, 19])
    plt.ylim([min(min(r_0), min(r_1)) - 2, max(max(r_0), max(r_1)) + 2])
    
    # Gráficos da autocorrelação
    plt.subplot(len(vtVar)*2+1, 2, 2*(ik)+4)
    plt.plot(l,rs_1)
    plt.xlim([0, 19])
    plt.ylim([min(rs_1) - 1, max(rs_1) + 1])
    plt.legend(['rs_1, Variância:' + str(vtVar[ik])])
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

## Parâmetros
Eb = 1                               # Energia dos pulsos (normalizada para 1)
Er0 = Eb                             # Média da saída do correlator 0
Er1 = 0                              # Média da saída do correlator 1
vtEbN0_dB = np.array([-10, 0, 10])   # Eb/N0s a simular em dB
vtEbN0 = 10**(vtEbN0_dB/10)          # Eb/N0s a simular em linear
vtVar = Eb*Eb/vtEbN0/2               # Variância na entrada do detector (= EN0/2)

for ik in range(len(vtEbN0_dB)):
    
    fig1 = plt.figure(figsize = (15, 20))
    
    dStd = np.sqrt(vtVar[ik])
    x = np.arange(-5*dStd-Eb,5*dStd + Eb, 0.001) # Eixo (5*variância)
    
    ## Cálculo da distribuição Gausiana
    vtr_0 = norm.pdf(x,Er0,dStd)                      # Valores de r0 (distribuição de probalilidade r0|s0)
    vtr_1 = norm.pdf(x,Er1,dStd)                      # Valores de r1 (distribuição de probalilidade r1|s0)
    
    ## Gráficos
    plt.subplot(len(vtEbN0_dB),1,ik+1)
    plt.plot(x,vtr_0,x,vtr_1)
    plt.title('Eb/No =' + str(vtEbN0_dB[ik]))

    plt.show()
```
