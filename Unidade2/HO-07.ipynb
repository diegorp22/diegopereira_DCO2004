```matlab
clc; clear all;close all;
%% DISTRIBUIO "Normal"
%% Parmetros da distribuio
vtMu = [0 -10 10];              % Valores de mdia da Gaussiana
vtVar = [1 5 10];               % Valores de varincia da Gaussiana
x = -20:0.1:20;
% Variando a mdia e plotando os grficos
fig = figure;
chLegend = [];
for ik=vtMu
    mu = ik;
    sigma = sqrt(vtVar(1));
    pd1 = makedist('Normal','mu',ik,'sigma',sigma);
    y1 = pdf(pd1,x);
    y2 = cdf(pd1,x);
    subplot(2,2,1);
    plot(x,y1);
    hold all;
    title('PDF com varincia fixa (0)');
    subplot(2,2,2);
    plot(x,y2);
    hold all;
    %
    title('CDF com varincia fixa (0)');
    chLegend = [chLegend; {['Mdia =  ' num2str(mu )]}];
end
subplot(2,2,1);
legend(chLegend);
subplot(2,2,2);
legend(chLegend);
chLegend = [];
% Variando a mdia e plotando os grficos
for ik=vtVar
    sigma = ik;
    mu = vtMu(1);
    pd2 = makedist('Normal','mu',mu,'sigma',ik);
    y1 = pdf(pd2,x);
    y2 = cdf(pd2,x);
    subplot(2,2,3);
    plot(x,y1);
    title('PDF com mdia fixa (1)');
    hold all;
    subplot(2,2,4);
    plot(x,y2);
    title('CDF com mdia fixa (1)');
    hold all;
    %
    chLegend = [chLegend; {['\sigma =  ' num2str(sigma)]}];
end
subplot(2,2,3);
legend(chLegend);
subplot(2,2,4);
legend(chLegend);
fig.PaperUnits = 'inches';
fig.PaperPosition = [0 0 12 6];
```
